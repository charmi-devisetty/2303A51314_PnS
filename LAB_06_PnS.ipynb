#A small business owner tracks the number of daily customers visiƟng their store, where each
#day receives between 0 and 10 visitors. The owner wants insights into the visit paƩerns.
# Generate a dataset represenƟng the number of customers visiƟng the store per day
#over N days.
# Compute and visualize the PMF and CDF to analyze the likelihood of receiving a certain
#number of customers on a given day.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generate_customer_data(N, min_visitors=0, max_visitors=10):
    return np.random.randint(min_visitors, max_visitors + 1, size=N)
```

```python
def compute_pmf(data):
    unique, counts = np.unique(data, return_counts=True)
    pmf = counts / len(data)
    return unique, pmf
```

```python
def compute_cdf(data):
    unique, counts = np.unique(data, return_counts=True)
    cdf = np.cumsum(counts) / len(data)
    return unique, cdf
```

```python
# Parameters
N = 365  # Simulating data for one year
data = generate_customer_data(N)
```

```python
# Compute PMF and CDF
x_pmf, y_pmf = compute_pmf(data)
x_cdf, y_cdf = compute_cdf(data)
```

```python
# Visualization
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(12, 5))
```

```python
# Plot PMF
plt.subplot(1, 2, 1)
plt.bar(x_pmf, y_pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
```

```python
# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x_cdf, y_cdf, marker='o', linestyle='-', color='red')
plt.xlabel("Number of Customers")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
```

```python
plt.tight_layout()
plt.show()
```

#A university professor esƟmates that 70% of students in a large class (N ≥ 100) will pass an
#exam. Suppose N students take the exam:
#i) What is the probability that exactly 60 students pass?
#ii) What is the probability that at least 75 students pass?
#iii) What is the expected number of students passing the exam?


```python
from scipy.stats import binom

# Given data
N = 100  # Number of students
p = 0.7  # Probability of passing

# (i) Probability that exactly 60 students pass
prob_60 = binom.pmf(60, N, p)
print(f"P(X = 60): {prob_60:.5f}")

# (ii) Probability that at least 75 students pass
prob_at_least_75 = 1 - binom.cdf(74, N, p)
print(f"P(X >= 75): {prob_at_least_75:.5f}")

# (iii) Expected number of students passing
expected_value = N * p
print(f"Expected number of students passing: {expected_value}")
```
